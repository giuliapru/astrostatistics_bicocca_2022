```python
import numpy as np
import matplotlib.pyplot as plt

x = []
N = 10000
a = 0.1
b = 100
for i in range(N):
    x.append(np.random.uniform(a, b))

x = np.array(x)
y = np.log10(x)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))

xaxis = np.linspace(a,b)
uniform_x = 1/(b-a)
ax[0].hist(x, density = True)
ax[0].plot(xaxis, uniform_x+0*xaxis)

xaxis1 = np.linspace(np.min(y), np.max(y)-0.1, 100)
log_dist = 10**(xaxis1)*np.log(10)/(b-a)
ax[1].hist(y, density = True)
ax[1].plot(xaxis1, log_dist)
```

```python
mean_y = np.mean(y)
log_mean_x = np.log10(np.mean(x))
#print()

median_y = np.median(y)
log_median_x = np.log10(np.median(x))
#print()
```
